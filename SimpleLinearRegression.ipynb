```python
#Pandas is the library for data manipulation and analysis 
#Numpy is the library for numberical operations
#matplotlib pyplot is the module for datavisualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading data set using read_csv function
datasets = pd.read_csv("Salary_Data.csv")
```

```python
#viewing the dataset
datasets
```

```python
# splitting input & output variables
independent = datasets[["YearsExperience"]] #Indepedent Variable,years of Experience, datatype float
dependent = datasets[["Salary"]] # Depedent Variable, Salary, datatype float
```

```python
#creating a scatter plot graph to see the input & output variables marked
plt.scatter(independent,dependent)
plt.xlabel('YearsExperience',fontsize=20)
plt.ylabel('Salary',fontsize=20)
plt.show()
```

```python
#this uses scikit-learn's train_test_split function to divide train test data & test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent,dependent, test_size = 1/3, random_state = 0)
```

```python
# this imports the linear regression & the fit formula from the scikit-learn function
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)#y=W*x1+b0 for this equation we got value for W and bo
```

```python
# Viewing the Weight(slope) and bias(error) value
weight=regressor.coef_
print("Weight of the model={}".format(weight))
bias=regressor.intercept_
print("Intercept of the model={}".format(bias))
```

```python

#evaluation Metrics
y_pred=regressor.predict(X_test)
```

```python
#calulation of evalution metrics
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#nearly 1 is good model, nearly 0 is poor model
#saving model
import pickle
filename="finalized_model.sav" #creating filename
pickle.dump(regressor,open(filename,'wb')) #uploading model into filename.wb-writeby, rb-readby
```

```python
# Load the previously saved model from the file 'finalized_model.sav'
# a prediction using the loaded model with input value 15
loaded_model=pickle.load(open("finalized_model.sav",'rb'))
result=loaded_model.predict([[15]]) #checking result for values
result
```
